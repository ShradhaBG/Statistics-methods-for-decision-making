```python
#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
#Load the data
df = pd.read_csv("austo_automobile.csv")
```

```python
##Checking the Types of the Data
print(df.dtypes)
```

```python
#checking for any missing/null values

df.isnull().sum()
```

```python
#treating null values

missing_values = df.isnull().sum()

print("Missing Values:")
print(missing_values)

#Treat missing values if needed

if missing_values.sum() > 0:


    numerical_columns = df.select_dtypes(include='number').columns
    df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

    categorical_columns = df.select_dtypes(exclude='number').columns
    df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

    print("\nMissing values treated successfully.")

# Check if missing values have been handled
print("\nMissing Values After Treatment:")
print(df.isnull().sum())
```

# Problem 1 - Bivariate Analysis
- Explore the relationship between all numerical variables - Explore the correlation between all numerical variables - Explore the relationship between categorical vs numerical variables

```python
#Exploring the relationship between all numerical variables:

numerical_var = ['Age', 'No_of_Dependents', 'Salary', 'Partner_salary', 'Total_salary', 'Price']

# Pairplot for numerical variables
sns.pairplot(df[numerical_var])
plt.show()
```

```python
#correlation between all numerical variables:

correlation_matrix = df[numerical_var].corr()

# Heatmap for correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# relationship between categorical vs numerical variables:

categorical_Var = ['Gender', 'Profession', 'Marital_status', 'Education', 'Personal_loan', 'House_loan', 'Partner_working', 'Make']

for cat_column in categorical_Var:
    for num_column in numerical_var:
        sns.boxplot(x=cat_column, y=num_column, data=df)
        plt.title(f'{cat_column} vs {num_column}')
        plt.show()
```

```python
import pandas as pd
data = pd.read_excel("godigt_cc_data.xlsx")
```

```python
print(data.shape[0])
print(data.columns)
print(data.dtypes)
```

```python
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
key_variables = [
    'high_networth', 'cc_active30', 'cc_active60', 'cc_active90',
    'Transactor_revolver', 'avg_spends_l3m', 'cc_limit',
    'T+1_month_activity', 'T+2_month_activity', 'T+3_month_activity',
    'T+6_month_activity', 'T+12_month_activity'
]

mapping_high_networth = {'A': 5, 'B': 4, 'C': 3, 'D': 2, 'E': 1}
mapping_transactor_revolver = {'T': 1, 'R': 0}
data_mapped = data[key_variables].copy()
data_mapped['high_networth'] = data_mapped['high_networth'].map(mapping_high_networth)
data_mapped['Transactor_revolver'] = data_mapped['Transactor_revolver'].map(mapping_transactor_revolver)


correlation_matrix = data_mapped.corr()


correlation_with_future_activity = correlation_matrix.loc[
    ['high_networth', 'cc_active30', 'cc_active60', 'cc_active90', 'Transactor_revolver', 'avg_spends_l3m', 'cc_limit'],
    ['T+1_month_activity', 'T+2_month_activity', 'T+3_month_activity', 'T+6_month_activity', 'T+12_month_activity']
]

correlation_with_future_activity
```

```python
summary_df = pd.DataFrame({
    "Variable": ["cc_active30", "cc_active60", "cc_active90", "avg_spends_l3m", "Transactor_revolver"],
    "Highest Correlation with Future Activity": [
        correlation_with_future_activity.loc["cc_active30"].max(),
        correlation_with_future_activity.loc["cc_active60"].max(),
        correlation_with_future_activity.loc["cc_active90"].max(),
        correlation_with_future_activity.loc["avg_spends_l3m"].max(),
        correlation_with_future_activity.loc["Transactor_revolver"].max()
    ]
})

summary_df.set_index("Variable", inplace=True)

summary_df
```
